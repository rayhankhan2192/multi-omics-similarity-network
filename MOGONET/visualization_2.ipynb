```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Example structure
data = [
    {"Dataset": "ROSMAP", "Method": "Cosine Similarity", "Accuracy": 0.825, "F1 weighted": 0.817, "F1 macro": 0.766},
    {"Dataset": "ROSMAP", "Method": "Cosine Distance", "Accuracy": 0.783, "F1 weighted": 0.772, "F1 macro": 0.700},
    {"Dataset": "ROSMAP", "Method": "RBF Similarity", "Accuracy": 0.821, "F1 weighted": 0.810, "F1 macro": 0.745},
    {"Dataset": "ROSMAP", "Method": "RBF Distance", "Accuracy": 0.764, "F1 weighted": 0.764, "F1 macro": 0.671},
    {"Dataset": "ROSMAP", "Method": "Hybrid Similarity", "Accuracy": 0.764, "F1 weighted": 0.753, "F1 macro": 0.685},
    {"Dataset": "ROSMAP", "Method": "Hybrid Distance", "Accuracy": 0.776, "F1 weighted": 0.766, "F1 macro": 0.693},
]


df = pd.DataFrame(data)

plt.figure(figsize=(6, 4))
sns.barplot(data=df, x="Method", y="Accuracy", hue="Dataset")
plt.title("Best Accuracy by Similarity Method")
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import pi
import matplotlib.ticker as ticker

df = pd.DataFrame(data)

# Radar chart setup
metrics = ["Accuracy", "F1 weighted", "F1 macro"]
num_vars = len(metrics)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # close the circle

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Plot each method
for _, row in df.iterrows():
    values = row[metrics].tolist()
    values += values[:1]
    ax.plot(angles, values, label=row["Method"])
    ax.fill(angles, values, alpha=0.1)

# Set up axis
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles[:-1]), metrics)

# Custom radial ticks from 0.7 to 1.0
custom_ticks = np.arange(0.6, 1.01, 0.1)
ax.set_ylim(0.6, 1.0)
ax.set_yticks(custom_ticks)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f"{x:.2f}"))
ax.set_rlabel_position(180 / num_vars)

plt.title("BRCA")
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.0))
plt.tight_layout()
plt.show()
```

```python
df = pd.DataFrame(data)

# Melt into long format
df_melt = df.melt(id_vars=["Method"], value_vars=["Accuracy", "F1 weighted", "F1 macro"],
                  var_name="Metric", value_name="Value")

# Plot Violin Plot
plt.figure(figsize=(9, 6))
sns.violinplot(data=df_melt, x="Metric", y="Value", inner="point", linewidth=1.2)

# Customize
plt.title("BRCA", fontsize=14)
plt.ylim(0.6, 1.0)
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Original data
data = [
    {"Dataset": "ROSMAP", "Method": "Cosine Similarity", "Accuracy": 0.825, "F1 weighted": 0.817, "F1 macro": 0.766},
    {"Dataset": "ROSMAP", "Method": "Cosine Distance", "Accuracy": 0.783, "F1 weighted": 0.772, "F1 macro": 0.700},
    {"Dataset": "ROSMAP", "Method": "RBF Similarity", "Accuracy": 0.821, "F1 weighted": 0.810, "F1 macro": 0.745},
    {"Dataset": "ROSMAP", "Method": "RBF Distance", "Accuracy": 0.764, "F1 weighted": 0.764, "F1 macro": 0.671},
    {"Dataset": "ROSMAP", "Method": "Hybrid Similarity", "Accuracy": 0.764, "F1 weighted": 0.753, "F1 macro": 0.685},
    {"Dataset": "ROSMAP", "Method": "Hybrid Distance", "Accuracy": 0.776, "F1 weighted": 0.766, "F1 macro": 0.693},
]


df = pd.DataFrame(data)

# Melt for grouped bar plot
df_melted = df.melt(id_vars="Method", value_vars=["Accuracy", "F1 weighted", "F1 macro"], var_name="Metric", value_name="Score")

plt.figure(figsize=(6, 4))
import seaborn as sns
sns.barplot(data=df_melted, x="Method", y="Score", hue="Metric")
plt.xticks(rotation=15)
plt.title("ROSMAP: Accuracy, F1, AUC by Method")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

methods = [d["Method"] for d in data]
metrics = ["Accuracy", "F1 weighted", "F1 macro"]
angles = np.linspace(0, 2 * np.pi, len(metrics), endpoint=False).tolist()
angles += angles[:1]

plt.figure(figsize=(8, 8))
for d in data:
    values = [d[m] for m in metrics]
    values += values[:1]  # Close the circle
    plt.polar(angles, values, label=d["Method"], linewidth=2)

plt.xticks(angles[:-1], metrics)
plt.title("ROSMAP: Metric Radar Chart per Method", y=1.1)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.tight_layout()
plt.show()
```

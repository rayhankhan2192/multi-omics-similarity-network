```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import pi
import matplotlib.ticker as ticker

# Original data
data = [
    {"Dataset": "ROSMAP", "Method": "Cosine Similarity", "Accuracy": 0.877, "F1": 0.876, "AUC": 0.902},
    {"Dataset": "ROSMAP", "Method": "Cosine Distance", "Accuracy": 0.821, "F1": 0.819, "AUC": 0.863},
    {"Dataset": "ROSMAP", "Method": "RBF Similarity", "Accuracy": 0.849, "F1": 0.846, "AUC": 0.892},
    {"Dataset": "ROSMAP", "Method": "RBF Distance", "Accuracy": 0.858, "F1": 0.860, "AUC": 0.871},
    {"Dataset": "ROSMAP", "Method": "Hybrid Similarity", "Accuracy": 0.849, "F1": 0.846, "AUC": 0.881},
    {"Dataset": "ROSMAP", "Method": "Hybrid Distance", "Accuracy": 0.830, "F1": 0.820, "AUC": 0.886},
]

df = pd.DataFrame(data)

# Radar chart setup
metrics = ['Accuracy', 'F1', 'AUC']
num_vars = len(metrics)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # close the circle

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Plot each method
for _, row in df.iterrows():
    values = row[metrics].tolist()
    values += values[:1]
    ax.plot(angles, values, label=row["Method"])
    ax.fill(angles, values, alpha=0.1)

# Set up axis
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles[:-1]), metrics)

# Custom radial ticks from 0.7 to 1.0
custom_ticks = np.arange(0.6, 1.01, 0.1)
ax.set_ylim(0.7, 1.0)
ax.set_yticks(custom_ticks)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f"{x:.2f}"))
ax.set_rlabel_position(180 / num_vars)

plt.title("ROSMAP")
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.0))
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data
data1 = [ 
    {"Dataset": "BRCA", "Method": "Cosine Similarity", "Accuracy": 0.825, "F1 weighted": 0.817, "F1 macro": 0.766},
    {"Dataset": "BRCA", "Method": "Cosine Distance", "Accuracy": 0.783, "F1 weighted": 0.772, "F1 macro": 0.700},
    {"Dataset": "BRCA", "Method": "RBF Similarity", "Accuracy": 0.821, "F1 weighted": 0.810, "F1 macro": 0.745},
    {"Dataset": "BRCA", "Method": "RBF Distance", "Accuracy": 0.764, "F1 weighted": 0.764, "F1 macro": 0.671},
    {"Dataset": "BRCA", "Method": "Hybrid Similarity", "Accuracy": 0.764, "F1 weighted": 0.753, "F1 macro": 0.685},
    {"Dataset": "BRCA", "Method": "Hybrid Distance", "Accuracy": 0.776, "F1 weighted": 0.766, "F1 macro": 0.693},
]

data2 = [ 
    {"Dataset": "ROSMAP", "Method": "Cosine Similarity", "Accuracy": 0.877, "F1": 0.876, "AUC": 0.902},
    {"Dataset": "ROSMAP", "Method": "Cosine Distance", "Accuracy": 0.821, "F1": 0.819, "AUC": 0.863},
    {"Dataset": "ROSMAP", "Method": "RBF Similarity", "Accuracy": 0.849, "F1": 0.846, "AUC": 0.892},
    {"Dataset": "ROSMAP", "Method": "RBF Distance", "Accuracy": 0.858, "F1": 0.860, "AUC": 0.871},
    {"Dataset": "ROSMAP", "Method": "Hybrid Similarity", "Accuracy": 0.849, "F1": 0.846, "AUC": 0.881},
    {"Dataset": "ROSMAP", "Method": "Hybrid Distance", "Accuracy": 0.830, "F1": 0.820, "AUC": 0.886},
]

# Prepare data
df1 = pd.DataFrame(data1).melt(id_vars="Method", 
                               value_vars=["Accuracy", "F1 weighted", "F1 macro"], 
                               var_name="Metric", value_name="Score")
df1["Source"] = "BRCA"

df2 = pd.DataFrame(data2).melt(id_vars="Method", 
                               value_vars=["Accuracy", "F1", "AUC"], 
                               var_name="Metric", value_name="Score")
df2["Source"] = "ROSMAP"

combined = pd.concat([df1, df2], ignore_index=True)
combined["Metric"] = combined["Metric"].astype(str)
combined["Method"] = combined["Method"].astype(str)

# Create combined label for better distinction
combined["Method_Source"] = combined["Method"] + " (" + combined["Source"] + ")"

# Plot
plt.figure(figsize=(16, 9))
sns.set_style("whitegrid")
barplot = sns.barplot(
    data=combined,
    x="Method",
    y="Score",
    hue="Metric",
    ci=None,
    palette="Set2"
)

# Add markers for Source (circle vs square)
markers = {"BRCA": "o", "ROSMAP": "s"}
for i, row in combined.iterrows():
    xloc = list(combined["Method"].unique()).index(row["Method"])
    hue_index = list(combined["Metric"].unique()).index(row["Metric"])
    xpos = xloc - 0.3 + hue_index * 0.2
    plt.plot(xpos, row["Score"], 
             marker=markers[row["Source"]],
             color='black',
             markersize=6,
             alpha=0.8)

# Annotate bar values
for bar in barplot.patches:
    height = bar.get_height()
    if height > 0:
        barplot.annotate(f'{height:.2f}', 
                         (bar.get_x() + bar.get_width() / 2, height),
                         ha='center', va='bottom', fontsize=8, color='black')

plt.title("Combined Metrics Comparison with Source Distinction", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.legend(title="Metric", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Original datasets
data1 = [ 
    {"Dataset": "ROSMAP", "Method": "Cosine Similarity", "Accuracy": 0.825, "F1 weighted": 0.817, "F1 macro": 0.766},
    {"Dataset": "ROSMAP", "Method": "Cosine Distance", "Accuracy": 0.783, "F1 weighted": 0.772, "F1 macro": 0.700},
    {"Dataset": "ROSMAP", "Method": "RBF Similarity", "Accuracy": 0.821, "F1 weighted": 0.810, "F1 macro": 0.745},
    {"Dataset": "ROSMAP", "Method": "RBF Distance", "Accuracy": 0.764, "F1 weighted": 0.764, "F1 macro": 0.671},
    {"Dataset": "ROSMAP", "Method": "Hybrid Similarity", "Accuracy": 0.764, "F1 weighted": 0.753, "F1 macro": 0.685},
    {"Dataset": "ROSMAP", "Method": "Hybrid Distance", "Accuracy": 0.776, "F1 weighted": 0.766, "F1 macro": 0.693},
]

data2 = [ 
    {"Dataset": "ROSMAP", "Method": "Cosine Similarity", "Accuracy": 0.877, "F1": 0.876, "AUC": 0.902},
    {"Dataset": "ROSMAP", "Method": "Cosine Distance", "Accuracy": 0.821, "F1": 0.819, "AUC": 0.863},
    {"Dataset": "ROSMAP", "Method": "RBF Similarity", "Accuracy": 0.849, "F1": 0.846, "AUC": 0.892},
    {"Dataset": "ROSMAP", "Method": "RBF Distance", "Accuracy": 0.858, "F1": 0.860, "AUC": 0.871},
    {"Dataset": "ROSMAP", "Method": "Hybrid Similarity", "Accuracy": 0.849, "F1": 0.846, "AUC": 0.881},
    {"Dataset": "ROSMAP", "Method": "Hybrid Distance", "Accuracy": 0.830, "F1": 0.820, "AUC": 0.886},
]

# Convert to DataFrame and melt
df1 = pd.DataFrame(data1).melt(id_vars="Method", value_vars=["Accuracy", "F1 weighted", "F1 macro"], 
                               var_name="Metric", value_name="Score")
df1["Source"] = "BRCA"

df2 = pd.DataFrame(data2).melt(id_vars="Method", value_vars=["Accuracy", "F1", "AUC"], 
                               var_name="Metric", value_name="Score")
df2["Source"] = "ROSMAP"

# Combine and clean
combined = pd.concat([df1, df2], ignore_index=True)
combined["Metric"] = combined["Metric"].astype(str).str.strip()
combined["Method"] = combined["Method"].astype(str)

# Use different colors for different sources
palette = {
    'BRCA': 'skyblue',
    'ROSMAP': 'salmon'
}

# Plot
plt.figure(figsize=(16, 9))
sns.barplot(
    data=combined, x="Method", y="Score",
    hue="Source", palette=palette,
    ci=None, edgecolor="black"
)

# Add dots for each metric type
sns.stripplot(
    data=combined, x="Method", y="Score",
    hue="Metric", dodge=True, marker="D",
    alpha=0.6, size=6, linewidth=0.5
)

plt.title("Comparison of Metrics Across Methods and Sources", fontsize=16)
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
g = sns.FacetGrid(combined, col="Metric", hue="Source", sharey=False, height=4, aspect=1.5)
g.map(sns.lineplot, "Method", "Score", marker="o")
g.map(sns.scatterplot, "Method", "Score", s=100, edgecolor="black")
g.add_legend()
for ax in g.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(45)
g.fig.subplots_adjust(top=0.85)
g.fig.suptitle("Facet Comparison of Metrics from data1 and data2")
plt.show()
```

```python
# Pivot for heatmaps
pivot1 = df1.pivot(index="Method", columns="Metric", values="Score")
pivot2 = df2.pivot(index="Method", columns="Metric", values="Score")

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.heatmap(pivot1, annot=True, fmt=".3f", cmap="Blues", ax=axes[0])
axes[0].set_title("data1 Metrics Heatmap")

sns.heatmap(pivot2, annot=True, fmt=".3f", cmap="Greens", ax=axes[1])
axes[1].set_title("data2 Metrics Heatmap")

plt.suptitle("Heatmap Comparison of Method Performance")
plt.tight_layout()
plt.show()
```
